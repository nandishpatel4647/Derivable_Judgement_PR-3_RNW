```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("health_dataset.csv")
df.head()
```

```python
print("H0: Smoking has no effect on Diabetes.")
print("H1: Smoking affects Diabetes.")

print("\nH0: Mean BMI of Male and Female is equal.")
print("H1: Mean BMI of Male and Female is different.")
```

```python
# Age CI
mean_age = df['age'].mean()
std_age = df['age'].std()
n = len(df)

ci_age = stats.t.interval(0.95, df=n-1, loc=mean_age, scale=std_age/np.sqrt(n))

print("95% CI for Age:", ci_age)

# Weight CI
mean_weight = df['weight'].mean()
std_weight = df['weight'].std()

ci_weight = stats.t.interval(0.95, df=n-1, loc=mean_weight, scale=std_weight/np.sqrt(n))

print("95% CI for Weight:", ci_weight)
```

```python
# Z-test for BMI mean (assume population mean = 25)
sample_mean = df['bmi'].mean()
sample_std = df['bmi'].std()
n = len(df)

z = (sample_mean - 25) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-value:", z)
print("P-value:", p_value)
```

```python
male_bmi = df[df['gender'] == 'Male']['bmi']
female_bmi = df[df['gender'] == 'Female']['bmi']

t_stat, p_val = stats.ttest_ind(male_bmi, female_bmi)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
cont_table = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = stats.chi2_contingency(cont_table)

print("Chi-square value:", chi2)
print("P-value:", p)
```

```python
groups = [group['bmi'].values for name, group in df.groupby('age_group')]

f_stat, p_val = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_val)
```

```python
covariance = np.cov(df['age'], df['bmi'])[0][1]
correlation = np.corrcoef(df['age'], df['bmi'])[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

```python
alpha = 0.05

def decision(p):
    return "Reject H0" if p < alpha else "Fail to Reject H0"

print("Z-test decision:", decision(p_value))
print("T-test decision:", decision(p_val))
print("Chi-square decision:", decision(p))
print("ANOVA decision:", decision(p_val))
```
